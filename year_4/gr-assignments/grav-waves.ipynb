```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./report.mplstyle')
```

```python
X = 2
t_max = 4
num_pulses =5
A=0.1

omega = np.pi
```

```python
plt.figure(figsize=(8,8))
for i in range(num_pulses):
    t_em = i
    t_vals = np.linspace(t_em, t_em+X, 100)
    x = (t_vals - t_em)+ A*np.cos(omega*(t_vals -X))
    y = (t_vals - t_em)+ A*np.cos(omega*(t_vals -X))
    plt.plot(y, t_vals, 'g', label="Pulse Path" if i == 0 else "")
plt.scatter([0],[0],color='blue' ,label = "Pulsar ")
plt.scatter([X],[t_max],color = 'red',label = "Observer")
plt.xlabel("x (Position)")
plt.ylabel("t (Time)")
plt.legend()

plt.show()
```
