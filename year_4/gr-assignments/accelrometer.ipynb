```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('report.mplstyle')
```

```python
data1 = np.loadtxt('accel-data1.csv', delimiter=',', skiprows=1)
data2 = np.loadtxt('accel-data2.csv', delimiter=',', skiprows=1)
plt.figure(figsize=(12, 8))
plt.subplot(2,1, 1)
plt.plot(data1[:,0], data1[:,1], label='phyphox App \nMean: 9.767 m/s^2',color = 'blue')
plt.legend()
plt.ylabel('Acceleration (m/s^2)')
plt.subplot(2,1, 2)
plt.plot(data2[:,0], data2[:,3]*-9.816, label='Physics Toolbox App \nMean: 9.754 m/s^2', color='red')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)' ,)
plt.legend(loc = 'lower left')
plt.savefig('acceleration.png')
plt.show()
```

```python
print(np.mean(data1[:,1]), np.mean(data2[:,3]*-9.816))
```
