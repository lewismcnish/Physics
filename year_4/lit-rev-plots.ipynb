```python
import numpy as np 
import matplotlib.pyplot as plt
import mplhep as hep
title_size = 24
axis_size = 21
tick_size = 17
annotate_size = 19
axis_size_3d = 34
hep.style.use(hep.style.ATLAS)
plt.rcParams["mathtext.fontset"] = "custom"
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["mathtext.rm"] = "Times New Roman"
plt.rcParams["mathtext.it"] = "Times New Roman:italic"
plt.rcParams["mathtext.bf"] = "Times New Roman:bold"
```

# Cross section

```python
data = np.loadtxt('cross-section.csv', delimiter=',', skiprows=1)
errors = np.loadtxt('cross-section-unc.csv', delimiter=',', skiprows=1)
energy = data[:,0]
tt_prod = data[:,1]
t_channel = data[:,2]
s_channel = data[:,3]
tw_channel = data[:,4]
errors_tt_prod = errors[:,1]
errors_t_channel = errors[:,3]
errors_s_channel = errors[:,5]
errors_tw_channel = errors[:,7]
tt_bar = r"$t\\bart production$"


plt.figure(figsize=(10,6))
plt.plot(energy, tt_prod, label=r'$t\bar{t}$ production')
plt.plot(energy, t_channel, label='t-channel')
plt.plot(energy, s_channel, label='s-channel')
plt.plot(energy, tw_channel, label='tW-channel')
plt.legend( fontsize=annotate_size)
plt.errorsbar(energy, tt_prod, yerr=errors_tt_prod, capsize=5)
plt.errorsbar(energy, t_channel, yerr=errors_t_channel, capsize=5)
plt.errorsbar(energy, s_channel, yerr=errors_s_channel, capsize=5)
plt.errorsbar(energy, tw_channel, yerr=errors_tw_channel, capsize=5)
plt.xlabel('Energy [TeV]', fontsize=axis_size)
plt.ylabel('Cross section [pb]', fontsize=axis_size)
plt.xticks(fontsize=axis_size)
plt.yticks(fontsize=axis_size)
plt.savefig('cross-section.pdf')
plt.show()
```

```python


r = np.arange(-1,1, 0.01)
y = -10*np.abs(r)**2 + 12*np.abs(r)**4


plt.figure(figsize=(6,6))
plt.plot(r, y)
plt.xlabel(r'$\phi$', fontsize=axis_size)
plt.ylabel(r'V$(\phi)$', fontsize=axis_size)

plt.xticks(fontsize=0)
plt.yticks(fontsize=0)
plt.xlim(0,1)
plt.annotate('0', xy=(0,0), xytext=(-0.05, -1), fontsize=annotate_size)
# plt.legend('Higgs Potential', fontsize=annotate_size)
plt.savefig('higgs-pot.pdf')

plt.show()
```

```python

from matplotlib.ticker import LinearLocator



x = np.arange(-1, 1, 0.01)
y = np.arange(-1, 1, 0.01)
x, y = np.meshgrid(x, y)
R = np.sqrt(x**2 + y**2)
z = -1*np.abs(R)**2 + 0.7*np.abs(R)**4


fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
# Plot the surface.

surf = ax.plot_surface(x, y, z, cmap='viridis',
                       linewidth=0, antialiased=False)


ax.set_zlim(0, 1)
ax.set_xlim(0, 1.6)
ax.set_ylim(-1.6, 0)


ax.xaxis.set_tick_params(labelsize=0)
ax.yaxis.set_tick_params(labelsize=0)
ax.zaxis.set_tick_params(labelsize=0)


ax.grid(False)

ax.xaxis.pane.fill = False   
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.zaxis.set_rotate_label(False) 
ax.xaxis.set_rotate_label(False)
ax.yaxis.set_rotate_label(False)
ax.set_xlabel('$\phi_{Re}$', fontsize=axis_size_3d)
ax.set_ylabel('$\phi_{Im}$', fontsize=axis_size_3d)
ax.set_zlabel('V($\phi$)', fontsize=axis_size_3d)
ax.xaxis.labelpad = 3
ax.yaxis.labelpad = 5
ax.zaxis.labelpad = 10


# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_box_aspect(None, zoom=0.95)
plt.savefig('higgs-pot-3d.pdf')
plt.show()
```
